# <font color="#37BFF0">1.3 Importação das principais bibliotecas python necessárias para análise descritiva de um DataSet



Principais Bibliotecas em Python para Análise de Dados:

Existem várias bibliotecas em Python que são amplamente utilizadas para análise descritiva de dados. Aqui estão algumas das principais bibliotecas para análise descritiva:

a) **NumPy**: NumPy é uma biblioteca fundamental para computação numérica em Python. Ela fornece estruturas de dados eficientes para trabalhar com arrays multidimensionais e funções matemáticas de alto desempenho. É comumente usada para realizar **cálculos estatísticos básicos, como média, mediana, desvio padrão**, entre outros.

b) **Pandas**: Pandas é uma biblioteca popular para manipulação e análise de dados. Ela oferece estruturas de dados poderosas, como DataFrame e Series, que **facilitam a leitura, a manipulação e a análise de dados tabulares**. O Pandas fornece uma ampla gama de funcionalidades para **estatísticas descritivas, incluindo sumarização estatística, agregação de dados, filtros, ordenação e muito mais**.

c) **Matplotlib**: Matplotlib é uma biblioteca de visualização de dados em Python. Ela oferece uma ampla **variedade de gráficos**, permitindo a criação de visualizações detalhadas e personalizadas. O Matplotlib é frequentemente usado para criar **histogramas, gráficos de barras, gráficos de dispersão, gráficos de linha e muitos outros tipos de gráficos para análise exploratória de dados**.

d) **Seaborn**: Seaborn é uma biblioteca de visualização de dados baseada no Matplotlib. Ela fornece uma interface de alto nível para **criar gráficos estatísticos atraentes e informativos**. O Seaborn é particularmente útil para criar **visualizações estatísticas complexas**, como gráficos de distribuição, gráficos de regressão, matrizes de correlação e muito mais.

e) **SciPy**: Biblioteca que oferece uma ampla variedade de funcionalidades estatísticas, como **testes de hipóteses, ajuste de distribuições e estatísticas descritivas mais avançadas**.

Essas são algumas das bibliotecas populares para análise estatística descritiva e criação de gráficos em Python. Dependendo das suas necessidades específicas, outras bibliotecas, como StatsModels, Plotly e ggplot, também podem ser úteis.

A função **df.apply()** é aplicada a um objeto DataFrame (representado pela variável df) e permite aplicar uma função personalizada a um ou mais eixos desse DataFrame.

A função df.apply() recebe uma função como argumento e a aplica elemento por elemento ao longo do eixo especificado. Ela permite que você execute operações personalizadas em cada valor individual do DataFrame ou em cada linha ou coluna, dependendo do eixo escolhido.

A sintaxe básica da função df.apply() é a seguinte:

df.apply(func, axis=0)

Onde:

func é a função personalizada que você deseja aplicar a cada elemento, linha ou coluna do DataFrame.
axis é o argumento opcional que indica ao Pandas se você deseja aplicar a função ao longo das colunas (axis=0), ao longo das linhas (axis=1) ou a ambos (axis=None).
A função personalizada (func) deve ser definida pelo usuário e pode ser uma função interna ou uma função definida pelo usuário. Ela pode executar qualquer operação desejada nos elementos do DataFrame ou nas linhas/colunas, como cálculos matemáticos, transformações de dados, limpeza de dados, entre outros.

## <font color="#37BFF0">Cálculo das medidas de variabilidade
---

As Medidas de variabilidade são:

Amplitude Total = AT (medida mais grosseira de variabilidade dos dados)

Variância = Desvio-Padrão ao quadrado

Desvio-padrão (mais utilizada) = Raiz quadrada da variância

Coeficiente de Variação = Desvio-padrão dividido pela média (medida adimencional). CV = Desvio-padrão/média (%) . O CV é uma medida de variabilidade muito utilizada também, assim como o desvio-padrão.

Desvio Interquartílico, (IQR) ou Interquartile Range = Quartil3 - Quartil1
